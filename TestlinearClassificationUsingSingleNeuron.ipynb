```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generate some linearly seprable data
np.random.seed(0)
num_points = 100
X = np.random.rand(num_points, 2)
y = (X[:,0]+ X[:, 1]>0).astype(int)

#add a bias term(column of ones) to the input data
X = np.hstack((np.ones((num_points, 1)), X))

#initialize weights randomly
weights = np.random.rand(3)

#define the step activation function
def step_function(x):
  return np.where(x >= 0, 1,0)

#trining parameter
learning_rate = 0.1
num_epochs = 1000

#training the single neuron
for epoch in range(num_epochs):
  for i in range(num_points):
    #compute the weighted sum z
    z = np.dot(X[i],weights)
    #apply the step activation function
    y_pred = step_function(z)
    #update the weights
    weights += learning_rate * (y[i] - y_pred) * X[i]

#define the decision boundary function
def decision_boundary(x):
  return -(weights[0] + weights[1] * x) / weights[2]

#plot the data and the decision boundary
plt.scatter(X[:,1], X[:,2], c=y, cmap='bwr')
x_vals = np.linspace(-3,3,100)
y_vals = decision_boundary(x_vals)
plt.plot(x_vals, y_vals, 'k')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```
